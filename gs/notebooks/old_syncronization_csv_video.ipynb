```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
def read_csv_ganci(file_path):
    df_ = pd.read_csv(
        file_path,
        header=6,
        sep=';',
    )

    df = pd.DataFrame()
    df['PAN'] = df_.index
    for i, c in enumerate(df_.columns):
        if i > 0:
            df[c] = df_.iloc[:, i - 1].values

    df = df.set_index(pd.to_datetime(df.TIMESTAMP, unit='ms'))
    del df['TIMESTAMP']
    df = df.sort_index()
    
    return df
```

```python
path = '../dati'
files = [os.path.join(path, f) for f in os.listdir(path=path) if f.startswith('data')]
files.sort()
files
```

```python
dfs = []
for f in files:
    df = read_csv_ganci(f)
    dfs.append(df)
    print(f)
    print(df.index[0])
    print(df.index[-1])
    print('-'*10)
```

```python
def agganciato_sganciato_plot(df: pd.DataFrame, col_value: str, gancio: str, shift: int = 0) -> None:
    df_ = df[[col_value, gancio, 'sec']].copy()
    map_sec_status = df_[['sec', gancio]]
    map_sec_status = dict(map_sec_status.itertuples(index=False))
    df_[gancio] = (df_['sec'] + shift).map(map_sec_status)
    df_ = df_.dropna()

    df_['groups'] = (df_[gancio] != df_[gancio].shift()).cumsum()
    df_ = df_.groupby(by='groups').agg(v=(col_value, lambda x: pd.Series(x.values, index=x.index).to_dict()))
    for i in range(df_.shape[0]):
        color = 'red' if i%2 == 0 else 'green'
        pd.Series(df_.v.iloc[i]).plot(color=color)
```

```python
agganciato_sganciato_plot(df, col_value='AxTM2', gancio='gancio_dx', shift=28)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.manifold import TSNE
from datetime import datetime
```

```python
cols1 = [c for c in df.columns if c.endswith('1')][1:]
```

```python
dataset = df[cols1][2:-2].rolling('1s').mean()
dataset = StandardScaler().fit_transform(dataset)
kmeans = KMeans(n_clusters=3, random_state=1926, n_init='auto')
predictions = kmeans.fit_predict(dataset)
silhouette_score(dataset, predictions)
```

```python
df_ = df[cols1][2:-2].rolling('1s').mean()
df_['predicions'] = predictions
```
